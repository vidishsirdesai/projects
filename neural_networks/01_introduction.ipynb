# Introduction
Neural Networks are a type of machine learning model inspired by the structure an function of the human brain. They are made up of interconnected nodes or atrificial neurons, that process and learn from the data. These networks are used for various tasks, including image and speech recognition, natural language processing and decision making.

# Popular Applications Of Neural Networks
### Google advertisments
Google engine makes use of responsive ads to show relevant ads. For example,
- Consider that, a search is run for "Data Science" or "Linear Regression Model" on Google's search engine. Ads related to Data Science are shown to the user.
- Consider that, a search is run for "Food Recepies" on Google's search engine. Ads related to food delivery or cooking websites will be shown to the user.

All of the above is achieved using Neural Networks.

### Text completion
The search suggestions or word completion is another example where Neural Networks are used. For example,
- If the words "What is gen" is typed into the Google's search bar, search suggestion are made with "What is generative AI?" on the top.
- If an email is being composed, for example, if the words "Let's meet" are typed, text completion may suggest "Let's meet soon ..." or "Let's meet tomorrow ...".

### Data compression
With the introduction of social media and streaming services, user data generation has exploded in recent years. Storing all of this data for millions of users is challenging.

Neural Networks are used to compress the data in order to store it in a server. Essentially, an encoder is used to compress the data (image or video) into a compressed format (e.g., `.zip`, `.tar`, `.pkl`, etc) and a decoder is used to decompress the data into its original format.

Magic eraser uses image segmentation which helps in isolating the pixels of objects which are to be removed from the image.

Here, each pixel is classified or segmented into various classes. For example,
- all pixels belonging to water will be classified as class 0,
- all pixels belonging to humans will be classified as class 1,
- all pixels belonging to sand will be classified as class 2,
- etc.

This is known as image segmentation.
