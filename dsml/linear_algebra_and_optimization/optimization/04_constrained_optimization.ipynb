# Objective
To find the gradient of the loss function, i.e.,

$\bar{w} = \bar{w} - \eta(\nabla\bar{w}L(\bar{w}, w_0))$

$w_0 = w_0 - \eta(\nabla w_0L(\bar{w}, w_0))$

# Optimization Problem
$\min (- \sum^n_{i = 1} (\frac{\bar{w}^T.\bar{x} + w_0}{||\bar{w}||}) * y_i)$

The smallest value of $||\bar{w}||$ is going to give the minimum value of the function.

If the value of $||\bar{w}||$ is kept as small as possible, the loss function can be minimized. Consider a hyperplane defined by $\bar{w}$ and $w_0$,



If $||\bar{w}|| = 1$, the length of $\bar{w}$ is reduced but the direction is not changed.

Why do this? If the derivative (gradient) of the loss function is to be found without reducing the magnitude, it is very difficult. Time taken is also high, because, $||\bar{w}|| = \sqrt{w_1^2 + w_2^2 + ...}$.

Differentiating the loss function with respect to $\bar{w}$ and $w_0$ is a complex task.

Hence there is need for simplification. One of the ways to simplify is by reducing $||\bar{w}||$ to 1.

The new loss function is,

$\argmin_{(\bar{w}, w_0)_{subject to ||\bar{w}|| = 1}} - \sum^n_{i = 1} (\frac{\bar{w}^T.\bar{x} + w_0}{||w||}) * y_i$

How to find the gradient of such problems?

# Lagrange Multiplier
Consider a dummy example to understand the underlying concept of Lagrange multiplier. Say for a given function, $f(x) = \frac{4}{x}$, which has to be maximized,

$\max f(x) = \frac{4}{x} \\
\text{subject to } g(x) = 8 - 4x$

Plot of $f(x)$ looks as follows,



The plot of $f(x)$ and $g(x)$ looks as follows,



At what value of $x$, $f(x)$ is max, such that $g(x)$ is also True? The point of intersection of both the functions is where $f(x)$ is maximum.

The whole point of this is that, while maximizing or minimizing a function, if a condition is given, the search space can be limited or reduced.

In the above figure, the slope at the point will be equal to the slope of the line which is tangent.

When a function $f(x)$ has to be maximized or minimized given a condition $g(x)$ at the point of intersection, the slope or gradient of $f(x)$,

$\nabla f(x) = \lambda \nabla g(x)$

In the figure above, $\lambda = 1$. But this will not be the case in reality or in all the situations.

$f(x) = \frac{4}{x}, f'(x) = - \frac{4}{x^2}$

$g(x) = 8 - 4x, g'(x) = -4$

$- \frac{4}{x^2} = \lambda (-4)$

$\lambda = \frac{1}{x^2} = x^{-2}$

Lagrange's method in classical mechanics utilizes the Lagrangian ($L$), which often combines the system's objective (usually minimizing or maximizing an action integral) with its constraints (limitations on motion) into a single equation called the Lagrange equation.

The equation is give as,

$L(\lambda, x) = f(x) + \lambda g(x) \text{ [Either maximize/ minimize]}$

Where,
- $L(\lambda, x)$ = Lagrange function.

Why should they be combined? This is done because the computer cannot be given 2 equations.

# Lagrange's Findings
1. The gradient of objective function is proportional to the constraint at the point of intersection. Problem: The computer cannot be fed with 2 equations. $\nabla f(x) = \lambda \nabla g(x)$.
2. Combine Lagrange's function, find the gradient, equate it to 0 to find the minima of the function. $L(\lambda, x) = f(x) + \lambda g(x)$.

# Applying Lagrange's Findings On The Objective
$\min L(\bar{w}, w_0) = - \sum^n_{i = 1} (\bar{w}^T.x + w_0) * y_i$

$\text{ such that}: ||w|| = 1$

Combining the above 2 equations using Lagrange's function,

$L(\bar{w}, w_0, \lambda) = - \sum^n_{i = 1} (\bar{w}^T.x + w_0) * y_i + \lambda(||\bar{w}|| - 1)$

Finding the gradient of the above,

$\bar{w} = 
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix} \\
\bar{w}^T . \bar{w} = 
\begin{bmatrix}
w_1 & w_2 & w_3
\end{bmatrix} . 
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
\\ \\
\therefore \bar{w}^T . \bar{w} = (w_1^2 + w_2^2 + w_3^2)$

and,

$||\bar{w}|| = \sqrt{w_1^2 + w_2^2 + w_3^2} \\
||\bar{w}|| = \sqrt{\bar{w}^T . \bar{w}}$

Therefore,

$L(\bar{w}, w_0, \lambda) = - \sum^n_{i = 1} (\bar{w}^T.x + w_0) * y_i + \lambda(\sqrt{\bar{w}^T . \bar{w}} - 1)$

$\nabla L(\bar{w}, w_0) = 
\begin{bmatrix}
\frac{}{\bar{w}}L(\bar{w}, w_0, \lambda) \\
\frac{}{\bar{w_0}}L(\bar{w}, w_0, \lambda)
\end{bmatrix}$

Hence,

$\frac{}{\bar{w}} L(\bar{w}, w_0, \lambda) = \sum^n_{i = 1} \frac{}{\bar{w}}(\bar{w}^T.x + w_0) * y_i + \lambda \frac{}{\bar{w}}(\sqrt{\bar{w}^T . \bar{w}} - 1)$

Solving the first part of the above equation produces the following as output,

$\sum^n_{i = 1} \frac{}{\bar{w}}(\bar{w}^T.x + w_0) * y_i = x. y_i$

Solving the second part of the above equation produces the following as output,

$\frac{}{\bar{w}}(\lambda .\sqrt{\bar{w}^T . \bar{w}} - \lambda ) = \frac{\lambda . \bar{w}}{\sqrt{\bar{w}^T . \bar{w}}}$

Finally,

$\nabla \bar{w} L(\bar{w}, w_0, \lambda) = - \sum^n_{i = 1} x . y_i + \frac{\lambda . \bar{w}}{\sqrt{\bar{w}^T . \bar{w}}}$

Now,

$\frac{}{w_0} L(\bar{w}, w_0, \lambda) = \sum^n_{i = 1} \frac{}{\bar{w}}(\bar{w}^T.x + w_0) * y_i + \lambda \frac{}{\bar{w}}(\sqrt{\bar{w}^T . \bar{w}} - 1)$

The above equation will reduce to,

$\frac{}{w_0} L(\bar{w}, w_0, \lambda) = \sum^n_{i = 1} \frac{}{w_0}w_0.y_i$

Finally,

$\nabla w_0 L(\bar{w}, w_0, \lambda) = - \sum^n_{i = 1} y_i$

Therefore the final equation of gradient descent will be,

$\bar{w} = \bar{w} - \eta(\nabla \bar{w} L(\bar{w}, w_0, \lambda)) = - \sum^n_{i = 1} x . y_i + \frac{\lambda . \bar{w}}{\sqrt{\bar{w}^T . \bar{w}}}$

$w_0 = w_0 - \eta(\nabla w_0 L(\bar{w}, w_0, \lambda)) =  - \sum^n_{i = 1} y_i$

# Types Of Gradient
# Classical gradient descent
Consider the following to be the data with 1 million rows,

| $x_1$ | $x_2$ | $x_3$ | $y$ |
| :-: | :-: | :-: | :-: |
|||||

Update rule for gradient descent is,

$\bar{w} = \bar{w} - \eta(- \sum^n_{i = 1} (x . y_i + \frac{\lambda . \bar{w}}{\sqrt{\bar{w}^T . \bar{w}}})) ... \text{eq *} \\
w_0 = w_0 - \eta( - \sum^n_{i = 1} y_i)$

Say that $\bar{w}$ and $w_0$ are assigned random values at the start,

$\bar{w} = 
\begin{bmatrix}
1 \\ 1 \\ 1
\end{bmatrix} \\
w_0 = 1$

If $\bar{w}$ has to be updated for one time, the entire dataset has to be iterated through fromm $i = 1 to n$ in order to find the summation of $\text{eq *}$.

This again has to be repeated for some m number of times.

If $m = 1000$ and $n = 1000000$, then $O(1000 * 1000000)$.

The above method of updating is called as classical or vanilla or batch gradient descent.

The solution to the time complexity is stochastic gradient descent

### Stochastic gradient fescent
In stochastic gradient descent, the entire data is not iterated through, instead one random data point is considered. Therefore,

$\bar{w_{new}} = \bar{w_{old}} - \eta(x . y_i + \frac{\lambda . \bar{w}}{\sqrt{\bar{w}^T . \bar{w}}}))$

$w_{0new} = w_{0old} - \eta(y_i)$

For each iteration of update, 1 random data point is considered. Therefore, the time complexity is reduced to O(1000) if total iterations are 1000.

But this could prove to be counter productive, as only limited number of data points are considered.

The research has suggested that even this method can give sufficiently good results,
- if the data is less in size.
- if the number of iterations are more.

The solution for this is mini-batch gradient descent.

### Mini-batch gradient descent
In Mini-batch gradient descent, m data points are considered, and $1 < m < n$. Meaning, a subset of the entire dataset is considered,

$\bar{w} = \bar{w} - \eta(- \sum^m_{i = 1} (x . y_i + \frac{\lambda . \bar{w}}{||w||}))$

$w_0 = w_0 - \eta( - \sum^m_{i = 1} y_i)$

The different types of gradient descents are,
- Batch.
- Stochastic.
- Mini-batch



### Epoch v. iteration
When entire dataset is iterated through once it is called as an epoch.

# Code For Gradient Descent

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
def show3d(x, y, w, w0):
    fig = plt.figure()
    ax = fig.add_subplot(projection = "3d")
    
    # plot the plane
    xx, yy = np.meshgrid(np.arange(-10, 10, 0.1), np.arange(10, 10, 0.1))
    z = (- w[0] * xx - w[1] * yy - w0) * 1. /w[2]
    ax.plot_surface(xx, yy, z, alpha = 0.4)

    # plot the dataset
    for m, l in [('o', -1), ('^', 1)]:
        xx = x[y == l, :]
        xs = xx[:, 0]
        ys = xx[:, 1]
        zs = xx[:, 2]
        ax.scatter(xs, ys, zs, marker=m)

    ax.set_xlabel("X Label")
    ax.set_ylabel("Y Label")
    ax.set_zlabel("Z Label")

    plt.show()
```

```python
def show2d(x, y, w, w0):
    line_x = np.linspace(-5, 10)
    line_y = (-w[0]*line_x - w0)/w[1]
    plt.plot(line_x, line_y) # show h-plane
    
    plt.scatter(x[:, 0], x[:, 1], c = y)
    plt.xlim([x[:, 0].min()-0.8, x[:, 0].max()+0.8])
    plt.ylim([x[:, 1].min()-0.8, x[:, 1].max()+0.8])
    plt.show()
```

```python
def get_gradient_w(x, y, w, w0, l):
    g = np.zeros_like(w)
    for xi, yi in zip(x,y):
        g -= yi * xi
    g += l * w / np.linalg.norm(w)
    return g
```

```python
def get_gradient_w0(x, y, w, w0, l):
    g = 0
    for xi, yi in zip(x,y):
        g -= yi
    return g
```

```python
def get_loss(x, y, w, w0, la):
    l = 0
    for xi, yi in zip(x, y):
        l -= ((np.dot(xi, w) + w0) * yi)
    l += la * (np.linalg.norm(w) - 1)
    return l
```

```python
def get_accuracy(x, y, w, w0):
    pl = []
    for xi,yi in zip(x, y):
        pll = np.sign(np.dot(w, xi) + w0)
        if pll == yi:
            pl.append(0)
        else:
            pl.append(1)
    pl = np.array(pl)
    
    plt.figure()
    plt.scatter(x[pl==0, 0], x[pl==0, 1], c="b")
    plt.scatter(x[pl==1, 0], x[pl==1, 1], c="r")
    plt.title("Accuracy")
    
    acc = 1 - (np.sum(pl)/len(pl))
    return acc * 100
```

```python
# create dataset
num_features = 3
x, y = datasets.make_blobs(
	n_samples=200, 
	centers=[[1]*num_features, 
	[8]*num_features], 
	n_features=num_features, 
	center_box=(0, 10), 
	cluster_std=1
)
y[y==0] = -1

# initial values for our classifier
w = np.array([-1/3]*num_features)
w0 = -7
l = 1000

# cell which performs one step of G.D.
if num_features == 2:
    show2d(x, y, w, w0)
    # plot_line(w, w0)
elif num_features == 3:
    show3d(x, y, w, w0)
    # plot_line(w, w0)
else:
    raise NotImplementedError("Can only do this for 2 or 3 dimensions")

print("Accuracy: ", get_accuracy(x, y, w, w0))
print("Loss: ", get_loss(x, y, w, w0, l))

# gradient update
w  = w  - l * get_gradient_w(x, y, w, w0, l)
w0 = w0 - l * get_gradient_w0(x, y, w, w0, l)

print("w:  ", w)
print("w0: ", w0)
```
