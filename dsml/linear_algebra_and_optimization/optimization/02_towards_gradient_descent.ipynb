The maximum value of $f(x) = 0$. The minimum value of $f(x) = -\infty$.

$x^* = \argmax_{x} -(x - 2)^2 \\
x^* = 2$.

Interest is to find out the location where the maxima lies, and it is not to find out the maxima itself.

# Differentiation: Geometric Picture
Consider,



Both, $y_1$ and $y_2$ can be replaced with $f(x_1)$ and $f(x_2)$,

$\therefore m = \tan\theta = \frac{f(x_2) - f(x_1)}{x_2 - x_1}$

Now consider,



This slope can be found even if the point $x_2$ is brought closer to $x_1$.

Let the point $x_2$ be brought closer to $x_1$ such that,

$x_2 = x_1 + \Delta x, \text{ Where } \Delta x = \text{ very very small}\\
x_2 = x_1 + h_{h -> 0}$

Now,

$m = \frac{f(x_2) - f(x_1)}{x_2 - x_1}$

$m = \frac{f(x_1 + h) - f(x_1)}{x_1 + h - x_1}$

$\therefore m = \frac{f(x_1 + h) - f(x_1)}{h}$

The derivative of $f(x)$,

$f'(x) = \lim_{h -> 0} \frac{f(x_1 + h) - f(x_1)}{h}$

Since the gap is small, an assumption can be made that the line of the slope touches the curve only once.

Note: Secant is a line that touches the curve more than once. Tangent is a line that touches the curve only once.

The derivative of a point is the angle the tangent makes at that point. Therefore,

"*Derivative of $f(x)$ at $x = a$, is the angle of the tangent at $x = a$*". It is denoted by $f'(a)$.

# How To Find The Derivative Of Any Function?
Using the *ab-initio* method.

Consider, $f(x) = x^2$.

Therefore, $f'(x) = \lim_{h -> 0} \frac{(x + h)^2 - x^2}{h} = 2x$.

The *ab-initio* method assumes that the derivative always exists. But in actuality, this is not the case. The derivative does not always exist for a function.

Hence, before calculating the derivative, differentiability has to be checked (find if the function is differentiable).

# Check For Differentiability
### Rules
1. For a function to be differentiable at point x, the $f(x)$ must be continuous at x. Every function that is continuous is not differentiable.
2. The derivative of the function $f'(x)$ must be continuous at x. LHL of $f'(x)$ = RHL of $f'(x)$ = $f'(x)$. Not all continuous functions are differentiable.

Derivative is also known as instantenous rate of change of $f(x)$ at $x = a$.

# Python Code To Find The Derivative Of A Function

```python
# code to find the derivative using the ab-initio method
def derivative(x1, fx):
    h = 0.0000001

    der = (fx(x1 + h) - fx(x1))/ h

    return der

func = lambda x: x ** 3
derivative(2, func)
```

```python
# code to plot any function and its derivative
import numpy as np
import matplotlib.pyplot as plt

def plot_func(func):
	x = np.linspace(-10, 10, 1000)
	y = [func(i) for i in x]

	plt.plot(x, y)
	plt.show()

func = lambda x: x ** 2
plot_func(func)
```

```python
def plot_func_der(func):
	x = np.linspace(-10, 10, 1000)
	y = [func(i) for i in x]
	deriv = [derivative(i, func) for i in x]

	plt.plot(x, y)
	plt.plot(x, deriv)
	plt.legend(["function", "derivative"])
	plt.show

func = lambda x: x ** 2
plot_func_der(func)
```

# Properties Of Derivatives
### Linearity
If $h(x) = f(x) + g(x)$, then $h'(x) = f'(x) + g'(x)$.

### Product rule
If $h(x) = f(x) * g(x)$, then $h'(x) = f'(x)g(x) + f(x)g'(x)$.

### Quotient rule
If $h(x) = \frac{f(x)}{g(x)}$, then $h'(x) = \frac{g(x).f'(x) - g'(x).f(x)}{g(x)^2}$.

### Chain rule
If $h(x) = f(g(x))$, then $h'(x) = f'(g(x)) * g'(x)$.

# How Are Derivatives Used For Optimization?
Consider,



What is the slope of the curve at 1, 2, 3, 4, and 5? The slope is 0 at all of these points.

At any point just before point 1, the slope will be negative. The angle will be more than $90^{\degree}$ with x-axis.

At any point just before point 2, the slope will be positive. And the point after 2 the slope will be negative.

For the points, 1, 3 and 5, the slope is changing from negative direction to positive direction. These are called as minima points.

For the points 2 and 4, the slope is changing from positive direction to negative direction. These are called as maxima points.

To summarize, valleys are minima, and peaks are maxima.

### How to find the candidate points for maxima and minima?
Find all the points where $f'(x)$ is zero. But this condition gives both maxima and minima. This segregation can be made by taking the derivative of point just before and after the point, where $f'(x) = 0$.

The logic from the above (direction of slope) will help in segregation there after.

But the above approach is cumbersome. A better approach is to find the second order derivative.

Second order derivatives are,
- If $f(x) = x^3$, 
- first order derivative is, $f'(x) = 3x^2$
- and second order derivative is, $f''(x) = 6x$.

Therefore,
- if $f''(x) > 0$, then minima.
- if $f''(x) < 0$, then maxima.
- if $f''(x) = 0$, then saddle point.

For $f''(x) > 0$ (minima), the rate of change of derivative is positive, hence it is minima.

Consider,



when there is no change in the slope before and after, the point is called a saddle point.
