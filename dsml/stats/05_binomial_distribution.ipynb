# Introduction
Bi means 2, for example, heads or tails, ON or OFF, Yes or No, 1 or 0, etc.

Any situation or process that has 2 outcomes is binomial in nature.

# What Is Binomial Distribution?


Binomial distribution is a statistical distribution that summarizes the probability that, a value will take one of the two independent values under a given set of parameters or assumptions. The underlying assumptions of binomial distribution are that there is only one outcome for each trial, each trial has the same probability of success and each trial is mutually exclusive or independent of one another.

The binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent and identical distribution (IID) Bernoulli's trials.

Each Bernoulli trail is a random experiment with 2 possible outcomes, success or failure.

The Binomial distribution is widely used on probability and statistics to model situations where an event has only 2 outcomes, such as, Success or Failure, Yes or No, Heads or Tails, etc.

# Key Characteristics
1. Number of trials ($n$): This represents the fixed number of trials or experiments. In each trial the event of interest occurs with a certain probability of success (usually denoted by $p$).
2. Probability of success ($p$): In each trial, there is a probability p of success, where $0 <= p <= 1$. The probability of failure is $(1 - p)$.
3. Independence: The trials are assumed to be independent. Meaning, the outcome of one trial does not affect the outcomes of other trials.
4. Discreteness: The number of success is a discrete random variable, meaning it can only take on whole numbers (0, 1, 2, ...).

# Example

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from scipy.stats import binom, geom
```

```python
# two trials: {SS, SF, FS, FF}
p = 0.1
x_vals = [0, 1, 2]
probs = [(1 - p) * (1 - p), p * (1 - p) + (1 - p) * p, p * p]
sns.barplot(x = x_vals, y = probs)
```

```python
# three trials : {FFF, FFS, FSF, FSS, SFF, SFS, SSF, SSS}
p = 0.1
x_vals = [0, 1, 2, 3]
probs = [
	(1 - p) * (1 - p) * (1 - p),
	(1 - p) * (1 - p) * p + (1 - p) * (1 - p) * p + (1 - p) * (1 - p) * p,
	(1 - p) * p * p + (1 - p) * p * p + (1 - p) * p * p,
	p * p * p
]
sns.barplot(x = x_vals, y = probs)
```

```python
# the above process of creating complex equations is simplified by the following two lines of code
probs_binom = binom.pmf(x_vals, n = 3, p = 0.1)
probs_binom
sns.barplot(x = x_vals, y = probs_binom)
```

```python
# n trials
p = 0.1 # left skewed
# p = 0.5 # gaussian 
# p = 0.9 # right skewed
n = 7
# n = 20
x_vals = np.arange(0, n + 1)
x_vals
probs_binom = binom.pmf(x_vals, n = n, p = p)
probs_binom
sns.barplot(x = x_vals, y = probs_binom)
```

# Bernoulli's Distribution
When only 1 trial is considered, it is called as the Bernoulli's distribution. The Bernoulli's distribution is discrete probability distribution representing 2 possible outcomes of random experiment, "success" and "failure".

### Key characteristics
1. A Bernoulli's trial is an experiment with 1 trial and 2 possible outcomes.
2. The probability of success (often described as $p$) remains constant from trial to trial. Similarly, the probability of failure is $(1 - p)$.
3. Each trial is considered or assumed to be independent. Meaning the outcome of 1 trial does not affect the outcomes of other trials.

# Probability Mass Function
A Probability Mass Function (PMF), also sometimes called as probability function or frequency function, is a mathematical tool used in probability and statistics specifically for discrete random variables. It describes the probability of each possible value that the random variable can take.

### Key points
- Discrete random variables: PMFs apply to situations where the variables can only take certain distinct values. For example, the number of times a 6 sided dice is rolled (1, 2, 3, 4, 5 or 6) is a discrete random variable.
- Probability for each value: The PMF assigns a probability value (between 0 and 1) to each possible value of the variable. The sum of all these probabilities much always be equal to 1.
- Notation: The PMF for a random variable $X$ is usually denoted by $p(x)$ or P(X = x), which represents the probability that X takes the value x.

### What is PMF?
If $X$ is a discrete random variable, then its range $R_X$ is a countable set. So, it is therefore possible to list the elements in $R_X$,

$R_X = \{x_1, x_2, x_3, ...\}$.

Note that, here, $x_1$, $x_2$, $x_3$, ... are possible values of the random variable $X$. While random variables are usually denoted by capital letters, to represent the numbers in the range, lowercase letters are usually used, such as, $x$, $y$, $z$, $x_1$, etc.

For a discrete random variable $X$, the interest is to know the probabilities of $X = x_k$. Note that, there the event $A = \{X = x_k\}$ is defined as the set of outcomes in the sample space $S$, for which the corresponding value of $X$ is equal to $x_k$. In particular,

$A = \{s \in S | X(s) = k\}$

The probabilities of events, $\{X = x_k\}$ are formally shown by the Probability Mass Function (PMF) of $X$.

### Definition of PMF
Let $X$ be a discrete random variable with range,

$R_X = \{x_1, x_2, x_3, ...\} \text{ (Finite or countably finite)}$.

The function,

$P_X(X_k) = P(X = X_k), \text{ for } k = 1, 2, 3, ...$,

is called as the Probability Mass Function (PMF) of $X$.

### Properties
- $0 <= P_X(X) <= 1, \forall X$,
- $\sum_{x\in R_X} P_X(X) = 1$, and for any set,
- $A \subset R_X, P(X\in A) = \sum_{x\in A} P_X(X)$.

# NOTE
The probability value ($p$) defines the type of distribution. Say if $p$ is low, then the distribution will be right skewed. If $p$ has a mid value, then the distribution will be Gaussian and if $p$ has a high value, then the distribution will be left skewed.
