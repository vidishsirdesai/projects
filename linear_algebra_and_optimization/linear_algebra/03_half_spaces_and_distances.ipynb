# Trigonometry Of Angles



- $\sin\theta = \frac{P}{H}$.
- $\cos\theta = \frac{B}{H}$.
- $\tan\theta = \frac{P}{B}$.

# Shifting Lines



How to move the line without changing the slope? Change $w_0$.

Instead, the origin itself can also be moved, the result of this algo gives a similar effect.

The origin can be moved up/ down/ left/ right depending on what the objective is,
- To shift the line by "a" units to the right, $w_1(x_1 - a) + w_2x_2 + w_0$.
- To shift the line by "a" units to the left, $w_1(x_1 + a) + w_2x_2 + w_0$.
- To shift the line by "a" units upwards, $w_1x_1 + w_2(x_2 - a) + w_0$.
- To shift the line by "a" units downwards, $w_1x_1 + w_2(x_2 + a) + w_0$.

Essentially,
- Shift up = decrease y.
- Shift down = increase y.
- Shift right = decrease x.
- Shift left = increase x.

# Distance Between Origin And Line
Consider,



Distance of the line from the origgin is given by the perpendicular drawn from the line connecting the origin to the point (in the above, it is $d_3$). The shortest distance is the perpendicular.

This is given by the L2 norm or the Euclidean distance.



$\cos\theta = \frac{B}{H} = \frac{d_3}{5}$

$\cos60 = \frac{d_3}{5}$

$d_3 = 5\cos60 =2.5$



$\therefore \cos\theta = \frac{d_3}{||x_0||} \text{ ... eq1}$

Equation of the line,

$w^T . x + w_0 = 0$

$w_1x_1 + w_2x_2 + w_0 = 0$

$w^T . x_0 + w_0 = 0$

$w^T . x_0 = -w_0$

Angle between $\bar{w}$ and $\bar{x}$,

$\cos\theta = \frac{w^T x_0}{||w||||x_0||} \text{ ... eq2}$

$\therefore \frac{d_3}{||x_0||} = \frac{w^T x_0}{||w||||x_0||}$

$d_3 = \frac{w^T x_0}{||w||}$

$d = \frac{w^T x_0}{||w||}$

$\because w^Tx_0 = - w_0$

$\therefore d = \frac{-w_0}{||w||}$

The above is the formula for distance between origin and line.

For the equation,

$ax + by + c = 0$

$distance = \frac{c}{\sqrt{a^2 + b^2}}$

The above is similar to,

$d = \frac{-w_0}{||w||}$

What is the significance of "-" in the above equation?

$\text{Coordinates of d} = 
\begin{bmatrix}
\frac{-w_0w_1}{w_1^2 + w_2^2} \\
\frac{-w_0w_2}{w_1^2 + w_2^2}
\end{bmatrix}$.

# Distance Between Point And Line
For the line, 

$ax + by + c = 0$

and the point,

$(x_1, y_1)$

d is given by,

$d = \frac{|ax_1 + by_1 + c|}{\sqrt{a^2 + b^2}}$

The linear algebra equivalent is,

$\frac{w_1x_1 + w_2x_2 + w_0}{||w||}$

$\therefore \frac{\bar{w}^T . \bar{x}}{||w||}$

Consider a point P whose coordinates are $(x_1', x_2')$,



$d = \frac{w^T . p}{||w||}$

# Code To Plot The Line $(w_1x + w_2y + w_0 = 0)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# w_vec = [w1, w2]
# w0 = <value>

def plot_line(w_vec, w_0):
    w1 = w_vec[0]
    w2 = w_vec[1]

    x_coords = np.linspace(-5, 5, 100)
    y_coords = (- w_0 - w1 * x_coords)/ w2

    plt.plot(x_coords, y_coords)

w_vec = [2, 3]
w_0 = 2

plot_line(w_vec, w_0)
```
