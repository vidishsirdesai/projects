# Law Of Total Probability
Consider that there are 3 regions with area,

$B_1 = 100Km^2, B_2 = 50Km^2, B_3 = 150Km^2$

and,

$\text{Total Area} = B_1+B_2+B_3 = 300Km^2$.

The same is the idea behind law of total probability, in which the area is replaced by probability of an event $A$.

The probability of event $A$, $P(A)$, can be calculated by partitioning the sample space $S$ and summing the probabilities of the portions of $A$ that lie within each partition.

For 2 events $A$ and $B$,

$P(A) = P(A \cap B) + P(A \cap B')$

using the definition of conditional probability, given by,

$P(A \cap B) = P(A|B) * P(B)$

Therefore,

$P(A) = P(A|B) P(B) + P(A|B') P(B')$.

A more generalized version of the above formula is, if $B_1$, $B_2$ and $B_3$ are a partition of the sample space $S$, then for any event $A$,

$P(A) = \sum_{i}^{n}P(A \cap B_i) = \sum_{i}^{n}P(A|B_i)P(B_i)$.

The above is the equation of law of total probability.

The idea behind the law of total probability can be visualized pictorially using a Venn diagram,

$A_1 = A \cap B_1, A_2 = A \cap B_2, A_3 = A \cap B_3$.



As it can be seen from the figure, $A_1$, $A_2$ and $A_3$ form a partition of the set $A$ and thus by the third axiom of probability,

$P(A) = P(A_1) + P(A_2) + P(A_3)$.
