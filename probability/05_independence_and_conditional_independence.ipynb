# Independence
Let $A$ be an event, where it rains tomorrow and suppose that,

$P(A) = \frac{1}{3}$.

Also suppose that a fair coin is tossed, let $B$ be the event that it lands with heads up. Therefore,

$P(B) = \frac{1}{2}$.

What would be the probability of $A$ given $B$?

$P(A|B) = P(A) = \frac{1}{3}$.

This is correct, because the result of coin toss does not have anything to do with the weather. Thus, no matter if B happens or not, the probability of A should not change.

This is an example of 2 independent events.

Therefore 2 events are independent, if one does not convey any information about the other. This can formally be defined as,

$\text{Two events A and B are indepndent if and only if }P(A \cap B) = P(A)P(B)$.

Reconciling this definition with what was mentioned earlier, $P(A | B) = P(A)$.

If 2 events are independent, then,

$P(A \cap B) = P(A)P(B)$.

So,

$P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{P(A)P(B)}{P(B)} = P(A)$.

Thus, if 2 events $A$ and $B$ are independent and $P(B) ≠ 0$, then $P(A | B) = P(A)$. To summarize, "*independence means, the probability of events can be multiplied to obtain the probability of their intersection*", or equivalently, "*independence means that conditional probability of one event given another is the same as the original (prior) probability*".

Sometimes the independence of 2 events is quite clear, because the 2 events seem not to have any physical interaction with each other (such as the 2 events discussed above). At other times, it is not as clear, and there might be a need to check if they satisfy the independence condition.

# Conditional Independence
Any concept that is defined for probability can also be extended to conditional probability.

2 events $A$ and $B$ are independent if,

$P(A \cap B) = P(A)P(B)$

or equivalently,

$P(A|B) = P(A)$.

This concept can be extended to conditionally independent events. The 2 events $A$ and $B$ are conditionally independent given an event $C$ with $P(C) > 0$, if,

$P(A \cap B|C) = P(A|C)P(B|C) \text{ - eq1}$

From the definition of conditional probability,

$P(A|B) = \frac{P(A \cap B)}{P(B)}$.

Where, $P(B) > 0$.

By conditioning on $C$,

$P(A|B, C) = \frac{P(A \cap B|C)}{P(B|C)}$

Where, $P(B | C)$, $P(C) ≠ 0$.

If $A$ and $B$ are conditionally independent given $C$,

$P(A|B, C) = \frac{P(A \cap B|C)}{P(B|C)}$

$P(A|B, C) = \frac{P(A|C) P(B|C)}{P(B|C)}$

Therefore,

$P(A|B, C) = P(A|C)\text{ - eq2}$

Equations 1 and 2 are equivalent statements of the definition of conditional independence.
